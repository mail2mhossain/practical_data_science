### Get the housing data

```python
import os
import tarfile
import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
fetch_housing_data()
```

```python
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path, na_values = missing_values)
```

```python
housing_df = load_housing_data()
```

## C. Data Normalization/Standardization (centering/scaling)

<b>Feature Scaling:</b> With few exceptions, Machine Learning algorithms don't perform well when the input numerical attributes have very different scales. Note that scaling the <b>target</b> values is generally not required.  

### Simple Feature Scaling


```python
auto_df["length"]
```

```python
auto_df["length"] = auto_df["length"]/auto_df["length"].max()
auto_df["length"]
```

### Min-Max (Normalization)


<b>After normalization, values are shifted and rescaled so that they end up ranging from 0 to 1. </b>

### Z-score (Standardization)


Unlike min-max scaling, standardization does not bound values to a specific range, which may be a problem for some algorithms (e.g., <b>neural networks</b> often expect an input value ranging 0 to 1)

## D. Data Binning

<b> Identifying categorical attribute </b>

If there are repetitive values in a column, which means that it is probably a categorical attribute. We can check it by using the value_counts method

```python
housing_df["ocean_proximity"].value_counts()
```

<b> Group a set of numerical values into a set of "bins" </b>

"price" is a feature range from 5,000 to 45,500 

(in order to have a better representation of price)



```python
bins = np.linspace(min(auto_df["price"]),max(auto_df["price"]),4)
```

```python
bins
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
auto_df["price_binned"] = pd.cut(auto_df["price"], bins, labels=group_names, include_lowest=True)
```

```python
auto_df[["price", "price_binned"]]
```
